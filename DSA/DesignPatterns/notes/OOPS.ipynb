## Basics of OOP

Object oriented programming is a paradigm based on wrapping state and behavior related to data into
special bundles called objects, which are constructed from a set of blueprints defined by a programmer
called classes.

Data stored inside object's fields is often referenced as state, and all the object's method define its
behavior.

### Dependency

Dependency indicates that the client element relies on the supplier element for some functionality or information.

ReportGenerator --->  DataFetcher
ReportGenerator is dependent on data fetcher.

```java

public class DataFetcher {
    public String fetchData() {
        return "Sample Data";
    }
}

public class ReportGenerator {
    private DataFetcher dataFetcher;

    // Constructor
    public ReportGenerator(DataFetcher dataFetcher) {
        this.dataFetcher = dataFetcher;
    }

    public void generateReport() {
        String data = dataFetcher.fetchData();
        // Generate report using the fetched data
        System.out.println("Report generated with data: " + data);
    }

    public static void main(String[] args) {
        DataFetcher dataFetcher = new DataFetcher();
        ReportGenerator reportGenerator = new ReportGenerator(dataFetcher);
        reportGenerator.generateReport();
    }
}
```

### Composition

 Composition represents a strong "has-a" relationship where one class is composed of one or more instances of another class. The composed object (part) cannot exist independently of the parent object (whole). When the parent object is destroyed, the child objects are also destroyed.

It's a way to show that one class (the whole) contains one or more instances of another class (the part), but the parts can exist independently of the whole.

### Introduction to design patterns

There are mainly 3 types of patterns:

- Creational patterns : Provide object creation mechanisms that increase flexibility and reuse of existing code.

- Structural patterns : Explain how to assemble objects and classes into larger structures, while keeping the structures flexible and efficient.

- Behavioral patterns : Take care of effective communication and assignment of responsiblities between objects.
