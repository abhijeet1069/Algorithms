# Design Principles

## Program to an interface not an implementation

Program to an interface, not an implementation. Depend on abstractions, not on concrete classes.

In left adding new foods will make code in Cat to fan out.

In right adding new foods will not make code in Cat to fan out. Also, we can see
food dependency has been inverted.

![Sample class](img/program_to_interface.png)

For code refer to designPrinciple.progamToInterface.bad,good

Inheritance also limits what functions can given out in public. Only functions of interface are visible to public. 
Everything else is hidden.

## Favor composition over inheritance

Inheritance is most suitable when there is a clear "is-a" relationship between the classes.
For ex : Dog is an Animal

Compostion gives has-a relationship
